## Read in data from local csv files

Note that the delimiter is ' ', not ','

```python
#url='http://www.metoffice.gov.uk/pub/data/weather/uk/climate/datasets/Sunshine/date/Scotland.txt'
frost = pd.read_csv('../weather/airfrost.csv', skiprows=7, skip_blank_lines=True, index_col='Year', delim_whitespace=True, parse_dates=True)
frost.loc[:'1980','JAN':'DEC']
```

```python
sunsh = pd.read_csv('../weather/sunshine.csv', skiprows=7, index_col='Year', delim_whitespace=True, parse_dates=True)
sunsh.loc['1961':'1980','JAN':'DEC']
```

### Concatenate columns from 2 Dataframes

Note: axis=1: columns, axis=0: rows

And also rename the columns, ready for producing a chart.

```python
frost_may = frost.loc['1961':'2016','ANN']
sunsh_may = sunsh.loc['1961':'2016','ANN']
sf = pd.concat([frost_may, sunsh_may], axis=1)
sf.columns=['Frost days', 'Sunshine hours']
sf.loc['2000':'2016']
```

Chart the data as 2 separate charts (different scales would obscure the frost days).

```python
sf.plot(kind='line', subplots=True)
```

Create a barchart of frost days in each month for 2016. To do this, first transpose the single row (2016) and the columns (of months)

```python
fp = frost.loc['2016','JAN':'DEC']
fs = fp.transpose()
fs.plot(kind='line', title='Frost days in each month of 2016')
```
